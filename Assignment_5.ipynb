# Question No One

# What is the northernmost airport in the United States?

```python
# To get the northernmost airport in United States, We will get the maximum value for Latitude in 
# the lat column in the Airports tabel.
```

```python
northernmost_airport = airports[airports['lat'] == airports['lat'].max()]
northernmost_airport.name
```

# Dillant Hopkins Airport is the northernmost airport in United States

```python
# To get the easternmost airport in United States, We will get the minimum value for Longtitude in 
# the lon column in the Airports tabel.
```

```python
easternmost_airport = airports[airports['lon'] == airports['lon'].min()]
easternmost_airport.name
```

# Adak Airport is the easternmost airport in United States

# On February 12th, 2013, which New York area airport had the windiest weather?

```python
# To New York area airport had the windiest weather, we should get the most value in the wind_speed
# column in the weather tabel
```

```python
weather.head()
```

```python
weather.origin.drop_duplicates()
```

```python
New_York_area_airport = weather[(weather.day == 12) & (weather.month == 2)
               & (weather.year == 2013) & (weather.origin == 'EWR')]
New_York_area_airport
```

```python
windiest_weather = New_York_area_airport[New_York_area_airport['wind_speed'] == 
                                         New_York_area_airport['wind_speed'].max()]
windiest_weather
```
